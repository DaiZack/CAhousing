```
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
data = pd.read_parquet('https://github.com/DaiZack/CAhousing/blob/master/housing_result.parquet?raw=true')
data.shape
# data.to_json('housing_result.json', orient='records')
```

```
data.head()
```

```
import re

p = re.compile(r'\bunited states\b|\bu\.?s\.\b|\bNew York\b|\bwashington\b|\bLos Angeles\b|\bChicago\b|\bHouston\b|\bPhiladelphia\b|\bPhoenix\b|\bSan Antonio\b|\bSan Diego\b|\bDallas\b|\bSan Jose\b|\bAustin\b|\bJacksonville\b|\bSan Francisco\b|\bIndianapolis\b|\bColumbus\b|\bFort Worth\b|\bCharlotte\b|\bSeattle\b|\bDenver\b|\bEl Paso\b'.lower())
pc = re.compile(r'\bcanada\b|\btoronto\b|\bedmonton\b|\bwinnipeg\b|\bcalgary\b|\bmontreal\\b|\bottawa\b|\bhamilton\b|\bvancouver\b')
```

```
data['united states'] = data.content.apply(lambda x: len(re.findall(p,x.lower()))>0 )
data['canada'] = data.content.apply(lambda x: len(re.findall(pc,x.lower()))>0)
```
